```python
import os
import random
import shutil
```

```python
source_1 = '/content/drive/MyDrive/UnderwaterImageEnhancement/dataset/train-set/raw/'
destination_1  = '/content/drive/MyDrive/UnderwaterImageEnhancement/dataset/test-set/raw/'


source_2 = '/content/drive/MyDrive/UnderwaterImageEnhancement/dataset/train-set/reference/'
destination_2  = '/content/drive/MyDrive/UnderwaterImageEnhancement/dataset/test-set/reference/'



files_1 = os.listdir(source_1)

#files_2 = os.listdir(source_2)

#20% images for the testing set
number_of_files = len(files_1) // 5
print(number_of_files)

for file_name in random.sample(files_1, number_of_files):

    try:
        # split the raw data into train set and test set
        shutil.move(source_1 + file_name, destination_1 + file_name)

        # split the reference data into train set and test set
        shutil.move(source_2 + file_name, destination_2 + file_name)


    except Exception as e:
            print(e)
```
